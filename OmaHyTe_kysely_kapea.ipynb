```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dat =pd.read_csv("https://raw.githubusercontent.com/kovanenirma/HyTe_IHME/master/Kopio_%20Kapea_esitysNan.csv")
dat[0:20]
```

```
dat2=dat[dat.Koodi=='K2']
dat2[0:10]
```

```
sns.set(style='darkgrid', rc={'figure.figsize':(16,8)})
ax = sns.barplot(x=dat.Kayttajanimi, y=dat2.Arvo, data=dat2)
```

```
dat3 = dat[dat.Kayttajanimi==222426]
dat3
```

```
sns.set(style='darkgrid', rc={'figure.figsize':(26,8)})
ax = sns.barplot(x=dat.Koodi, y=dat3.Arvo, data=dat3)
```

```
fiilis = dat[dat.Kysymys=='Naen lahes paivittain ihmisia kasvokkain']
teema = dat[dat.Kysymys=='Liikuin eilen tavoitteiden mukaisesti']

Fiilis = pd.Series(fiilis.Arvo)
Teema = pd.Series(teema.Arvo)

from numpy.polynomial.polynomial import polyfit
b, m = polyfit(Fiilis, Teema, 1)
plt.plot(Fiilis, Teema, '.')
plt.plot(Fiilis, b + m * Fiilis, '-')
plt.title("Teema vs. Fiilis")
plt.show()
```
